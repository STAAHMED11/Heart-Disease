```python

A0=['Male','Female']
B0=[] 
C0=[]
for a in A0:
    x=data.loc[(data['Sex']==a),:].loc[data['HeartDisease']=='Yes'].count()['HeartDisease']
    y=data.loc[(data['Sex'] ==a),].count()['HeartDisease']
    B0+=[x/y]
    C0+=[x]
px.bar(A0,B0,color=C0)
```

```python
D={}
L=['Smoking','AlcoholDrinking','Stroke','DiffWalking','PhysicalActivity','Asthma','KidneyDisease','SkinCancer']
for l in L:
    D[l]=[0,0]
Z=['No','Yes']
```

```python

for l in L:
    for i in Z:
        x=data.loc[(data[l]==i),:].loc[data['HeartDisease']=='Yes'].HeartDisease.count()
        y=data.loc[(data[l]==i),].count()['HeartDisease']
        r=Z.index(i)
        D[l][r]=x/y
print(D)
```

```python
fig = make_subplots(rows=2, cols=4, subplot_titles=L)
B=[None]*8
r=0
for i in range(1,3):
    for j in range(1,5):
        B[r]=(i,j)
        r+=1
r=0
for l in L:
    fig.add_trace(go.Bar(x=['No','Yes'], y=D[l],name=l), row=B[r][0], col=B[r][1])
    r+=1

fig.update_layout(height=1200, width=1500, title_text="effect for some factors on heart disease")
fig.show()
```

```python
# Check the class distribution
print("Original class distribution:")
print(data['HeartDisease'].value_counts())
print("\nClass distribution (%):")
print(data['HeartDisease'].value_counts(normalize=True) * 100)
```

```python
# Downsample the majority class (No) to balance the dataset
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = data[data['HeartDisease'] == 'No']
df_minority = data[data['HeartDisease'] == 'Yes']

# Downsample majority class to match minority class size
df_majority_downsampled = resample(df_majority,replace=False,n_samples=len(df_minority),random_state=42)

# Combine minority class with downsampled majority class
data_balanced = pd.concat([df_majority_downsampled, df_minority])

# Shuffle the dataset
data_balanced = data_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print("Balanced class distribution:")
print(data_balanced['HeartDisease'].value_counts())
print(f"\nOriginal dataset size: {len(data)}")
print(f"Balanced dataset size: {len(data_balanced)}")
```

```python
y=data_balanced.iloc[:,0]
X=data_balanced.iloc[:,1:]
y=y.replace('Yes',1)
y=y.replace('No',0) 
```

```python
X=X.replace('Male',B0[0])
X=X.replace('Female',B0[1])
```

```python
for a in range(len(A1)):
    X['AgeCategory']=X['AgeCategory'].replace(A1[a],B1[a])
for a in range(len(A2)):
    X['Race']=X['Race'].replace(A2[a],B2[a])
for a in range(len(A3)):
    X['GenHealth']=X['GenHealth'].replace(A3[a],B3[a])
for a in range(len(A4)):
    X['PhysicalHealth']=X['PhysicalHealth'].replace(A4[a],B4[a])
for a in range(len(A5)):
    X['SleepTime']=X['SleepTime'].replace(A5[a],B5[a])
for a in range(len(A6)):
    X['MentalHealth']=X['MentalHealth'].replace(A6[a],B6[a])         
for a in range(len(A)):
    X['Diabetic']=X['Diabetic'].replace(A[a],B[a])
for l in L:
    for a in range(len(Z)):
        X[l]=X[l].replace(Z[a],D[l][a])
```

```python
from sklearn.model_selection import train_test_split
# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

```

```python
y_train=y_train.replace('Yes',1)
y_train=y_train.replace('No',0)
y_train
```

```python
y_test=y_test.replace('Yes',1)
y_test=y_test.replace('No',0)
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost
accuracy_list=[]
recall_list=[]
```

```python
# logistic regression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
log_reg_pred = log_reg.predict(X_test)
log_reg_acc = metrics.accuracy_score(y_test, log_reg_pred)
log_reg_recall = metrics.recall_score(y_test, log_reg_pred)
accuracy_list.append(100*log_reg_acc)
recall_list.append(log_reg_recall)

print("logistic regression model accuracy(in %):", log_reg_acc*100)
print("logistic regression model recall:", log_reg_recall)
```

```python
# Decision Tree Classifier


dt_clf = DecisionTreeClassifier(max_leaf_nodes=3, random_state=0, criterion='entropy')
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
dt_acc = metrics.accuracy_score(y_test, dt_pred)
dt_recall = metrics.recall_score(y_test, dt_pred)
accuracy_list.append(100*dt_acc)
recall_list.append(dt_recall)


print('Decision Tree Classifier accuracy (in%): ', dt_acc*100)
print('Decision Tree Classifier recall: ', dt_recall)
```

```python
# K Neighbors Classifier

kn_clf = KNeighborsClassifier(n_neighbors=6)
kn_clf.fit(X_train, y_train)
kn_pred = kn_clf.predict(X_test)
kn_acc = metrics.accuracy_score(y_test, kn_pred)
kn_recall = metrics.recall_score(y_test, kn_pred)
accuracy_list.append(100*kn_acc)
recall_list.append(kn_recall)
print('K Neighbors Classifier accuracy:', kn_acc)

print('K Neighbors Classifier recall:', kn_recall)
```

```python
# GradientBoostingClassifier
gradientboost_clf = GradientBoostingClassifier(max_depth=2, random_state=1)
gradientboost_clf.fit(X_train,y_train)
gradientboost_pred = gradientboost_clf.predict(X_test)
gradientboost_acc = metrics.accuracy_score(y_test, gradientboost_pred)
gradientboost_recall = metrics.recall_score(y_test, gradientboost_pred)
accuracy_list.append(100*gradientboost_acc)
recall_list.append(gradientboost_recall)
print('GradientBoosting accuracy:', gradientboost_acc)

print('GradientBoosting recall:', gradientboost_recall)
```

```python

# xgbrf classifier
xgb_clf = xgboost.XGBRFClassifier(max_depth=3, random_state=1)
xgb_clf.fit(X_train,y_train)
xgb_pred = xgb_clf.predict(X_test)
xgb_acc = metrics.accuracy_score(y_test, xgb_pred)
xgb_recall = metrics.recall_score(y_test, xgb_pred)
accuracy_list.append(100*xgb_acc)
recall_list.append(xgb_recall)

print('XGBOOST accuracy:', xgb_acc)
print('XGBOOST recall:', xgb_recall)
```

```python

gnb = GaussianNB()
gnb.fit(X_train, y_train)
n_pred = gnb.predict(X_test)
n_acc = metrics.accuracy_score(y_test, n_pred)
n_recall = metrics.recall_score(y_test, n_pred)
accuracy_list.append(100*n_acc)
recall_list.append(n_recall)

print('GaussianNB accuracy:', n_acc)
print('GaussianNB recall:', n_recall)
```

```python
# Define parameters: these will need to be tuned to prevent overfitting and underfitting
params = {
    "n_estimators": 90,  # Number of trees in the forest
    "max_depth": 15,  # Max depth of the tree
    "min_samples_split": 12,  # Min number of samples required to split a node
    "min_samples_leaf": 6,  # Min number of samples required at a leaf node
    "ccp_alpha": 0,  # Cost complexity parameter for pruning
    "random_state": 123,
}
r_clf = RandomForestRegressor(**params)
r_clf = r_clf.fit(X_train, y_train)
r_pred =r_clf.predict(X_test)
r_pred=np.array(r_pred)
r_pred=np.round(r_pred)
r_pred=r_pred.astype(int)
r_acc =metrics.accuracy_score(y_test, r_pred)
r_recall = metrics.recall_score(y_test, r_pred)
accuracy_list.append(100*r_acc)
recall_list.append(r_recall)


print('Random Forest Regressor accuracy (in%): ', r_acc*100)
print('Random Forest Regressor recall: ', r_recall)
```

```python

model_list=['LogisticRegression','DecisionTreeClassifier','K Neighbors Classifier','GradientBoostingClassifier','XGBOOST','GaussianNB','Random Forest','SGDClassifier']

# Create comparison dataframe
comparison_df = pd.DataFrame({
    'Model': model_list * 2,
    'Score': accuracy_list + [r*100 for r in recall_list],
    'Metric': ['Accuracy']*len(model_list) + ['Recall']*len(model_list)
})

# Create grouped bar chart
fig = px.bar(comparison_df, x='Score', y='Model', color='Metric', barmode='group', orientation='h',title='Model Performance Comparison: Accuracy vs Recall',
             labels={'Score': 'Score (%)', 'Model': 'Model'},color_discrete_map={'Accuracy': '#636EFA', 'Recall': '#EF553B'})

fig.update_layout(height=600, width=1000)
fig.show()
```
